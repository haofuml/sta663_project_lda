```python
from sta663_project_lda.preprocessing.gen_toydata import toydata_generator
toydata_generator()
```

```python
import numpy as np
toydata_mat = np.load('./data/toydata_mat.npy')
toydata_voc = np.load('./data/toydata_voc.npy')
```

```python
svi_params = {
    'K':2, # topic_num
    'alpha0':0.5, # alpha_prior
    'gamma0':0.5, # gamma_prior
    'MB':20, # minibatch_size
    'kappa':0.5, # learning_rate=(tau0+itr)^(-kappa)
    'tau0':256,
    'eps':1e-3, # convergence criteria for local updates
}
```

```python
from sta663_project_lda.algorithms.lda_svi import LDASVI
from sta663_project_lda.algorithms.lda_gibbs import LDAGibbs
from sta663_project_lda.visualization.demo_topics import topic_viz
```

```python
lda = LDAGibbs('./data/toydata_mat.npy', ntopics=2)
lda.run()
topic_viz(lda.phi.T,toydata_voc,topk=5)
```

```python
lda = LDASVI('./data/toydata_mat.npy', **svi_params)
gamma_svi = lda.train_numpy(epoch=50, seed=0, printcycle=10)
topic_viz(gamma_svi, toydata_voc, topk=5)
```
